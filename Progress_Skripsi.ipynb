```python
import re
import emoji

def CleanTweetText(text:str) -> str:

  emojiToText = emoji.demojize(text,delimiters=('', '')).replace("_"," ")
  removeWebsites = re.sub(r'https?:\/\/.*[\r\n]*', '',emojiToText, flags=re.MULTILINE)
  removeMentions = re.sub(r'@\w+', " ",removeWebsites)
  removeHash = removeMentions.replace(r'#', " ")
  splitJoined = re.sub(r"(#?[A-Z][a-z]+)", r" \1", removeHash)
  lowercase = splitJoined.lower()
  removeNewLinerChar = lowercase.replace(r'\n', " ")
  removeNewLinerChar = removeNewLinerChar.replace(r'\r', " ")
  removeNumber = re.sub(r"[0-9\n]", " ", removeNewLinerChar)
  removePunctuation = removeNumber.translate(str.maketrans('', ''))

  # removeNonEnglishWords = " ".join(w for w in removePunctuation.split()
  #                                  if wordnet.synsets(w))
  removeSingleLetters = " ".join([word for word in removePunctuation.split(" ") if len(word)>=2])

  removeEmptySpaces = re.sub(r'[ ]{2,}', " ",removeSingleLetters)
  removeOuterSpaces = removeEmptySpaces.strip()

  return removeNumber

# print(CleanTweetText(dataset.full_text.values))
dataclean = CleanTweetText(dataset.full_text)

dataclean
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
```

```python
from typing import Tuple

def getNltkSentiments(row:pd.Series)->Tuple[np.float64]:
  text = str(row)
  score:dict = SentimentIntensityAnalyzer().polarity_scores(text)
  neg:np.float64 = score['neg']
  pos:np.float64 = score['pos']

  return neg, pos

for data in dataclean:
  print(getNltkSentiments(data))
```
